```python
from named_pipe import NamedPipe
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# C#과 테스트

```python
n = NamedPipe()
```

```python
len(n.read_byte())
```

```python
idx = 1
while True:
    n.read_byte()
    img_arr = n.get_img_arr()
    cv2.imwrite(f'{str(idx)}.bmp', img_arr)
    idx += 1
```

```python
plt.imshow(img_arr, cmap='gray')
```

```python
cv2.imwrite('result.bmp', img_arr)
```

```python
origin_img = cv2.imread('img.bmp', cv2.IMREAD_GRAYSCALE)
origin_img.shape
```

```python
result = img_arr-origin_img
np.unique(result)
```
